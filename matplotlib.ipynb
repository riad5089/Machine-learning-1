```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
#=np.arange(len(company))
y=np.arange(len(company))
y
```

```python
plt.xticks(y,company)

#plt.xticks(xpos,company)
plt.bar(y-0.2,revenue,width=0.4,label="Revenue")
plt.bar(y+0.2,profit,width=0.4, label="profit")
plt.legend()
plt.ylabel("Revenue")
plt.title("Stock of tech")
```

```python
plt.yticks(y,company)

#plt.xticks(xpos,company)
plt.barh(y-0.2,revenue,label="Revenue")
plt.barh(y+0.2,profit,label="profit")
plt.legend()
plt.ylabel("Revenue")
plt.title("Stock of tech")
```
